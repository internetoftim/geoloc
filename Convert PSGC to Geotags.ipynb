Sample of Location Data

Notes:
- pass the address to google api
- start_index, stop_index - for batch processing file
- rand_cutoff - random sampling of locations (percentage/probability, use all when negative)
- Google API has a daily query limit (2000 queries)
- get **key** from google developer account https://console.developers.google.com/apis/library/geocoding-backend.googleapis.com/?q=Google%20Maps%20Geocoding%20API&project=geoloc-187507


```python
i=0
start_index = 0
stop_index = 5
rand_cutoff = -1.0
url = 'https://maps.googleapis.com/maps/api/geocode/json'
import csv
import random
fieldnames = ['key','psgc', 'raw_address','raw_barangay','raw_town','raw_region','lat','long',
              'raw_result'
]

random.seed(a=500)

with open('psgc_geotags.csv', 'w') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter='|', quotechar='"', quoting=csv.QUOTE_NONE)
    spamwriter.writerow(fieldnames)
    for row in (temp_file.itertuples()):
        if(i<start_index): 
            i+=1
            continue
        if(i<stop_index):
            i+=1
            if(random.random()>rand_cutoff):
                temp_address =  row.Barangay + ' ' + row.Town + ' ' + (row.Region) + ' Philippines'
                print(temp_address)
                params = {'sensor': 'false', 'address': temp_address,'key':'AIzaSyCFhIDu6WDuqItFMTP7Mpb5axkLIT2R2LA'}
                r = requests.get(url, params=params)
                results = r.json()['results']
                time.sleep ((50.0 + random.random()*20)/ 1000.0)
                #print(results)
                #print(row._11)
                try:
                    location = results[0]['geometry']['location']
                    spamwriter.writerow([i,row.Index,temp_address,
                                         row.Barangay, row.Town,row.Region,
                                         str(location['lat']),str(location['lng']),
                                         results
                                    ] )
                except:
                    pass

                continue
            else:
                print(row.Region)
                continue
        else:
            break


        
```
